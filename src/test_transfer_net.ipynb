```python
import torch
import torch.nn as nn

from utils import display_image
from transfer_net import TransferNet
from PIL import Image
from torchvision import transforms
```

```python
image = Image.open('../data/test_data/dog.jpg')
image = transforms.Compose([transforms.Resize((500, 500)),transforms.ToTensor()])(image).unsqueeze(0)
display_image(image.squeeze(), 'Input image')

transfer_net = TransferNet()
params_count = sum(p.numel() for p in transfer_net.parameters() if p.requires_grad)
print('Tunable params:', params_count)

output = transfer_net(image)

display_image(output.detach().squeeze(), 'Output image')
```
